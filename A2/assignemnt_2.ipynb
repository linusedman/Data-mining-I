```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv("./www.csv", sep='\t')
```

```python
num_rows, num_columns = df.shape

print(f"Number of rows: {num_rows} \t Num columns: {num_columns-2}")
```

```python
#Check for outliers, missing values and duplicates

# check null values
nullValues = df.loc[:, df.isnull().any()]
print('nullvalues',nullValues)


# drop duplicates
print('nr of rows before dropping dups', len(df))
df = df.drop_duplicates('Query', keep='first')
df.reset_index(drop=True, inplace=True)
# print('nr of rows before after dups', len(df))

num_rows, num_columns = df.shape


print(f"Number of rows: {num_rows} \t Number of columns: {num_columns-2}")
```

```python
import pyfpgrowth

associations = df['Query'].apply(lambda x: x.split()).tolist()

associations = [list(set(sublist)) for sublist in associations]

num_records = len(associations)

print(len(associations))
print(associations)
```

```python
sigma = 100
min_support = sigma / num_records
print(f"Minimum support: {min_support}")

# the function find_frequent_patterns takes SIGMA as second parameter
patterns = pyfpgrowth.find_frequent_patterns(associations, sigma)
print(patterns)
```

```python
# number of frequent itemsets found
num_frequent_itemsets = len(patterns)

# maximum size of frequent itemsets
max_itemset_size = max(len(itemset) for itemset in patterns)
print(f'Number of frequent itemsets: {num_frequent_itemsets}')
print(f'Maximum size of frequent itemsets: {max_itemset_size}')


#most frequent keyword
keyword = max(patterns, key=patterns.get)
amount = patterns[keyword]
print(keyword, amount)
```

```python
# the support of an itemset is the fraction of records containing the items in the itemset
support = {key: value / len(df) for key, value in patterns.items()}
print(support)
```

```python
min_support_values = [0.001, 0.003, 0.004, 0.005, 0.008, 0.01]

for min_support in min_support_values:
    # the second parameter of the function 'find_frequent_patterns' is the support
    # count, NOT the min-support, so you have to calculate it
    support_count = min_support * num_records
    patterns = pyfpgrowth.find_frequent_patterns(associations, support_count)
    support = {key: value / len(df) for key, value in patterns.items()}
    print(f"Minimum support: {min_support}, Support count: {support_count:.0f}")
    print(f'Found {len(patterns)} patterns')
    print('Patterns', patterns)
    print('Support', support)
    print()
```

```python
# Set minimum support and confidence
min_support = 0.0007
support_count = min_support * num_records
confidence = 0.8

# Find frequent patterns
patterns = pyfpgrowth.find_frequent_patterns(associations, support_count)

# Generate association rules
rules = pyfpgrowth.generate_association_rules(patterns, confidence)

# Initialize a list to store rules with confidence and lift
rule_rankings = []

# Loop through the generated rules
for rule, values in rules.items():
    antecedent = rule  # Rule antecedent
    consequent = values[0]  # Rule consequent
    rule_confidence = values[1]  # Rule confidence
    
    # Calculate support for antecedent (X) and consequent (Y)
    support_X = len([assoc for assoc in associations if set(antecedent).issubset(assoc)])
    support_Y = len([assoc for assoc in associations if set(consequent).issubset(assoc)])
    
    # Calculate lift: (confidence / (support_Y / total number of transactions))
    lift = rule_confidence / (support_Y / num_records)
    
    # Store rule, confidence, and lift
    rule_rankings.append((antecedent, consequent, rule_confidence, lift))

# Sort rules by confidence (descending) and lift (descending)
rule_rankings.sort(key=lambda x: (-x[2], -x[3]))  # Sort by confidence first, then lift

# Print the top rules along with their confidence and lift
print(f"{'Rank':<5} {'Antecedent -> Consequent':<40} {'Confidence':<12} {'Lift':<10}")
for idx, (antecedent, consequent, rule_confidence, lift) in enumerate(rule_rankings):
    rule_str = f"{antecedent} -> {consequent}"
    print(f"{idx+1:<5} {rule_str:<40} {rule_confidence:<12.3f} {lift:<10.3f}")
```
